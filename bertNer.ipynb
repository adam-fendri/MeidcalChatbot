```python
import pandas as pd
from sklearn.model_selection import train_test_split
from transformers import RobertaTokenizerFast, RobertaForTokenClassification, Trainer, TrainingArguments
import torch
from sklearn.metrics import precision_recall_fscore_support, accuracy_score
import numpy as np

# Assuming `df` is your DataFrame with 'Sentence' and 'Tags'
df = merged_df

# Initialize the tokenizer for RoBERTa-large
tokenizer = RobertaTokenizerFast.from_pretrained('roberta-large')

def tokenize_and_align_labels(sentences, tags):
    tokenized_inputs = tokenizer(sentences, truncation=True, padding=True, is_split_into_words=False, return_tensors="pt")

    labels_all = []

    for i, sentence in enumerate(sentences):
        word_ids = tokenized_inputs.word_ids(batch_index=i)
        labels = tags[i].split()

        # Initialize label_ids with -100 (ignored index)
        label_ids = [-100] * len(tokenized_inputs['input_ids'][i])

        # Align labels with tokens
        for idx, word_id in enumerate(word_ids):
            if word_id is not None:  # Skip special tokens like [CLS] and [SEP]
                if word_id < len(labels):  # Check bounds
                    label_ids[idx] = label_to_id(labels[word_id])

        labels_all.append(label_ids)

    tokenized_inputs['labels'] = torch.tensor(labels_all)
    return tokenized_inputs

def label_to_id(label):
    label_map = {'O': 0, 'B-WEIGHT': 1, 'I-WEIGHT': 2, 'B-HEIGHT': 3, 'I-HEIGHT': 4, 'B-AGE': 5, 'I-AGE': 6}
    return label_map.get(label, -100)  # -100 is used for ignored index

# Split the data before tokenization
train_texts, test_texts, train_labels, test_labels = train_test_split(
    df['Sentence'].tolist(), df['Tags'].tolist(), test_size=0.2, random_state=42
)

# Tokenize and align labels for train and test sets
train_tokenized_inputs = tokenize_and_align_labels(train_texts, train_labels)
test_tokenized_inputs = tokenize_and_align_labels(test_texts, test_labels)

# Convert to PyTorch Dataset
class NERDataset(torch.utils.data.Dataset):
    def __init__(self, tokenized_inputs):
        self.tokenized_inputs = tokenized_inputs

    def __len__(self):
        return len(self.tokenized_inputs['input_ids'])

    def __getitem__(self, idx):
        item = {key: val[idx] for key, val in self.tokenized_inputs.items()}
        return item

train_dataset = NERDataset(train_tokenized_inputs)
test_dataset = NERDataset(test_tokenized_inputs)

# Define label map and get number of labels
label_map = {'O': 0, 'B-WEIGHT': 1, 'I-WEIGHT': 2, 'B-HEIGHT': 3, 'I-HEIGHT': 4, 'B-AGE': 5, 'I-AGE': 6}
num_labels = len(label_map)

# Load the RoBERTa-large model for token classification
model = RobertaForTokenClassification.from_pretrained('roberta-large', num_labels=num_labels)

# Define training arguments with optimization
training_args = TrainingArguments(
    output_dir='./results',               # output directory
    num_train_epochs=20,                  # number of training epochs
    per_device_train_batch_size=8,        # batch size for training (optimize for hardware)
    per_device_eval_batch_size=8,         # batch size for evaluation
    warmup_steps=200,                     # reduced number of warmup steps for learning rate scheduler
    weight_decay=0.01,                    # strength of weight decay
    logging_dir='./logs',                 # directory for storing logs
    logging_steps=20,
    evaluation_strategy='epoch',          # evaluate at the end of each epoch
    save_strategy='epoch',                # save model at the end of each epoch
    learning_rate=3e-5,                   # slightly lower learning rate for fine-tuning
    save_total_limit=2,                   # limit the number of saved checkpoints
    load_best_model_at_end=True,          # load the best model at the end of training
    metric_for_best_model='f1'            # use F1 as the metric to determine the best model
)

# Define compute metrics function for NER
def compute_metrics(pred):
    labels = pred.label_ids
    preds = pred.predictions.argmax(-1)

    # Flatten the predictions and labels
    labels_flat = labels.flatten()
    preds_flat = preds.flatten()

    # Only compute metrics for non-special tokens (ignore padding, CLS, SEP)
    mask = labels_flat != -100

    labels_flat = labels_flat[mask]
    preds_flat = preds_flat[mask]

    precision, recall, f1, _ = precision_recall_fscore_support(labels_flat, preds_flat, average='weighted')
    acc = accuracy_score(labels_flat, preds_flat)

    return {
        'accuracy': acc,
        'f1': f1,
        'precision': precision,
        'recall': recall
    }

# Initialize Trainer with metric computation
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    compute_metrics=compute_metrics
)

# Train the model
trainer.train()

# Evaluate the model
eval_results = trainer.evaluate()
print(eval_results)

# Save the model
model.save_pretrained('./ner_model_roberta_large')
tokenizer.save_pretrained('./ner_model_roberta_large')
```

```python
import torch
from transformers import RobertaTokenizerFast, RobertaForTokenClassification
import json

# Load the model and tokenizer
#model = RobertaForTokenClassification.from_pretrained('./ner_model_roberta_large')
#tokenizer = RobertaTokenizerFast.from_pretrained('./ner_model_roberta_large')

# Define your sentences for testing
test_sentences = [
    "I weigh 70 kg, my height is 175 cm.",
    "My height is 140 centimeters, I am 12 yrs old.",
    "With a weight of 25 kg, I am 8 yrs old.",
]

# Tokenize the input sentences
inputs = tokenizer(test_sentences, return_tensors="pt", padding=True, truncation=True, is_split_into_words=False)

# Get predictions from the model
with torch.no_grad():
    outputs = model(**inputs)

# Convert the model outputs to predicted labels
predictions = torch.argmax(outputs.logits, dim=-1)

# Define the label map
id_to_label = {0: 'O', 1: 'B-WEIGHT', 2: 'I-WEIGHT', 3: 'B-HEIGHT', 4: 'I-HEIGHT', 5: 'B-AGE', 6: 'I-AGE'}

# Extract predicted labels and the corresponding tokens
predicted_labels = []
for i, sentence in enumerate(test_sentences):
    tokens = tokenizer.tokenize(sentence)
    pred_labels = [id_to_label[pred.item()] for pred in predictions[i]]
    predicted_labels.append(list(zip(tokens, pred_labels)))

# Improved Entity Extraction Logic
def extract_entities(predicted_labels):
    entities = {"age": "", "height": "", "weight": ""}
    current_entity = None
    current_value = []

    for token, label in predicted_labels:
        if token.startswith("Ġ"):  # Handle space
            token = token[1:]  # Remove the special character for space

        if label.startswith("B-"):
            if current_entity:
                # Join the tokens for the previous entity
                entities[current_entity.lower()] = " ".join(current_value).strip()
            current_entity = label[2:]  # Start new entity
            current_value = [token]
        elif label.startswith("I-") and current_entity:
            current_value.append(token)  # Continue the entity
        else:
            if current_entity:
                # Finalize the current entity
                entities[current_entity.lower()] = " ".join(current_value).strip()
            current_entity = None
            current_value = []

    # Ensure the last entity is added
    if current_entity:
        entities[current_entity.lower()] = " ".join(current_value).strip()

    return entities

# Extract and print only the entities
for sentence, pred_labels in zip(test_sentences, predicted_labels):
    entities = extract_entities(pred_labels)
    print(f"Sentence: {sentence}")
    print("Entities:", json.dumps(entities, indent=4))
    print()
```
